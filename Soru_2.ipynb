```python
a=500
extreme=0
for x in range(0,a,100):
    if x%100==0 and x%400!=0:
        extreme+=1
print(extreme)
aa=(a-1) % 4
aaa=(((a-1)-aa)/4)-extreme
ilk_gun=-1 + (aaa)*2 + (a-aaa)*1 # Ayların ilk günüyle yılın ilk günü arasındaki fark için -1 eklenmiştir.
ayin_ilk_günü = [0,32,60,91,121,152,182,213,244,274,305,335]
artık_ayin_ilk_günü=[0,32,61,92,122,153,183,214,245,275,306,336]
for i in range(500,4001):
    if i%4 ==0:
        for j in range(0,11):
            if (artık_ayin_ilk_günü[j]+ilk_gun) % 7 ==0:
                tarih=str(1)+"/"+str(j+1)+"/"+str(i)
                print(tarih)
        if i%100==0 and i%400!=0 :
            ilk_gun+=1 #100'e ve 400'e tam bölünüp 4000'e tam bölünmeyen yıllar
        elif i%100==0 and i%400==0 and i%4000==0:
            ilk_gun+=1
        else:
            ilk_gun+=2 # 4 e bölünen yıllar 366 gün. 366 dan 7 nin tam katlarını çıkarırsanız bu yıllarda bir önceki yıla göre 2 gün kayma olur.
    else:
        for j in range(0,11):
            if (ayin_ilk_günü[j]+ilk_gun) % 7 ==0:
                tarih=str(1)+"/"+str(j+1)+"/"+str(i)
                print(tarih)
        ilk_gun+=1 # 4 e bölünmeyen yıllar 365 gün olduğu için bu yıllarda kaymalar önceki yıla göre 1 gün oluyor. 
                
```
