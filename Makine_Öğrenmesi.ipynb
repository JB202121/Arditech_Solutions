```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn.model_selection import validation_curve
from sklearn.model_selection import learning_curve
import numpy as np
a=np.arange(1,100.1,0.1)
b=a
x=np.zeros((len(a),2))
x[:,0]=a
x[:,1]=b
y=np.zeros((len(a)*len(a),1))
xx=np.zeros((len(a)*len(a),2))
k=0
for i in range(0,(len(a))):
    for ii in range (0,(len(b))):
        xx[k,0]=a[i]
        xx[k,1]=b[ii]
        y[k,0]=a[i]*b[ii]
        k=k+1
clf = tree.DecisionTreeRegressor()
clf=clf.fit(xx,y)
train_sizes, train_scores, valid_scores = learning_curve(clf, xx, y, cv=3)
plt.plot(train_scores)
#Learning linear regresörlerle iyi değildi fakat decision tree regresör kullanımını gördüm denedim fakat çok iyi sonuç verdi. 
#Neden bu kadar iyi çalıştı anlayamadım.
#Bu konuya belirli bir süre baktıktan sonra decision tree tanımlanırken eğer default olarak bırakılırsa (max depth) belirlenmezse
#En doğru sonuca ulaşana kadar devam edip bir nevi karar ağacını büyütüyormuş. Eğer max depth manuel olarak elle verilir ve
#küçük bir değer seçilirse accuracy bozulabilir. Training Accuracy mükemmel olmasına rağmen validasyonun bozuk olma sebebi
#buradan kaynaklanıyor. Amacımız maksimum doğruluk olduğu için max depth belirlemeyip default bırakıp en iyi sonucu aldım.
plt.title('Learning Accuracy') 
plt.figure()
plt.plot(valid_scores)
plt.title('Cross Validation Results')
result=clf.predict([[5.1,3.4],[4,3.5]])
print(result)
```
